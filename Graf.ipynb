```python
import matplotlib.pyplot as plt
import matplotlib.patheffects as pa
```

```python
import pandas as pd
```

```python
excel = pd.read_csv('finanzas2.csv',sep=';')
```

```python
excel
```

```python
excel['Fecha'] = pd.to_datetime(excel['Fecha'], dayfirst=True)
```

```python
excel.head()
```

```python
plt.style.use("ggplot")
fig,ax=plt.subplots(figsize = (15,8))
ax.plot(excel['Elon Mosk'],label='Elon Mosk',lw=2,color='#EC4D37',
        path_effects = [pa.SimpleLineShadow(),pa.Normal()])

ax.plot(excel['Bill Gate'],label='Bill Gate',lw=2,color='Blue',linestyle='-.',
        path_effects = [pa.SimpleLineShadow(),pa.Normal()])

ax.legend(shadow=True,prop = {'family':'Serif','size':15})
ax.set_ylabel('USD',fontname='Serif',fontsize=20)
ax.set_xlabel('Time',fontname='Serif',fontsize=20)
ax.set_title('Elon vs Gate',fontname='Serif',fontsize=25)

plt.show()
```
