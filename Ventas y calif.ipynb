```python
import matplotlib.pyplot as plt
import matplotlib.patheffects as pa
import pandas as pd
from matplotlib.ticker import FuncFormatter
```

```python
excel=pd.read_csv('compra+curso.csv', sep=';')
```

```python
excel
```

```python
excel.head()
```

```python
grafico= excel[['Compras_del_curso','calificaciones']]
```

```python
grafico.head()
```

```python
plt.style.use("seaborn-whitegrid")
fig,ax =plt.subplots(figsize=(15,8))
ax.plot(grafico.Compras_del_curso,grafico.calificaciones,marker='o',linestyle='',markersize=9,
        color='#F0A07C',markeredgecolor='#4A274F',markeredgewidth = 1.5)

ax.set_ylabel('Calificacion del curso',fontname='fantasy',fontsize=20)
ax.set_xlabel('Compras del curso',fontname='fantasy',fontsize=20)
ax.set_title('Ventas y Calificaciones',fontname='fantasy',fontsize=35, va= 'bottom')

def formato(x,pos):
    return '%.0f mil compras'%(x/1000)

ax.set_xticklabels(list(ax.get_xticks()),fontname='fantasy',fontsize=10)
ax.xaxis.set_major_formatter(FuncFormatter(formato))

ax.set_yticklabels(list(ax.get_yticks()),fontname='fantasy',fontsize=10)
plt.show();
```
